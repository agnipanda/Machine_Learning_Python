# Categorical Plot
### Dataset: 'tips'
The 'tips' dataset is a built-in dataset in 'Seaborn'.

```python
# import seaborn
# seaborn is a visualizing package
# built-over matplotlib
import seaborn as sns

# %matplotlib is exclusive to 'jupyter-notebook'
%matplotlib inline

# load 'tips' dataset
tips = sns.load_dataset('tips')
```

```python
# import the 'numpy' package
import numpy as np
```

```python
# display the first 7 rows of the 'tips' dataset
tips.head(7)
```

### Bar Plot
A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent.
A bar graph shows comparisons among discrete categories. One axis of the chart shows the specific categories being compared, and the other axis represents a measured value. Some bar graphs present bars clustered in groups of more than one, showing the values of more than one measured variable.

For more... https://en.wikipedia.org/wiki/Bar_chart

```python
# plotting a barplot to find the relation between 'sex' and 'total_bill'
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std) 

''' For parameter, 'estimator' try using 'np.sum', 'np.std', 'np.mean' '''
```

```python
# Similar to barplot but counts the number of the repetitions
# for a single column in a dataset
sns.countplot(x='sex',data=tips)

''' For parameter, 'x' try using 'sex', 'smoker', 'day', 'time', 'size '''
```

### Box Plot

A box plot is a method for graphically depicting groups of numerical data through their quartiles. 

Box plots may also have lines extending vertically from the boxes (whiskers) indicating variability outside the upper and lower quartiles, hence the terms box-and-whisker plot and box-and-whisker diagram. Outliers may be plotted as individual points.

A quartile is a type of quantile. The first quartile (Q1) is defined as the middle number between the smallest number and the median of the data set. The second quartile (Q2) is the median of the data. The third quartile (Q3) is the middle value between the median and the highest value of the data set.

For more... https://en.wikipedia.org/wiki/Box_plot

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# Here 'hue' categorizes the boxplot based on if the individual is a smoker or not
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

### Violin Plot
A violin plot is a method of plotting numeric data. It is similar to box plot with a rotated kernel density plot on each side.
Violin plots are similar to box plots, except that they also show the probability density of the data at different values.
A violin plot has four layers. The outer shape represents all possible results, with thickness indicating how common. (Thus the thickest section represents the mode average.) The next layer inside represents the values that occur 95% of the time. The next layer (if it exists) inside represents the values that occur 50% of the time. The central dot represents the median average value.

For more...https://en.wikipedia.org/wiki/Violin_plot

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

### Strip Plot
 Strip plot is a scatterplot where one variable is categorical.

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter = True,hue="sex", dodge=True)
''' "jitter" is used when we have many point in a single category and they overlap '''
```

### Swarm Plot
Swarm Plot is a categorical scatterplot with non-overlapping points.

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

### Factor Plot

The default plot that is shown is a point plot, but other seaborn categorical plots can be chosen with the kind parameter, including box plots, violin plots, bar plots, or strip plots.

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```
