```python
# import the necessary packages
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

iris = sns.load_dataset('iris')
iris.head()
```

## Pair Plot
Pair-plot plots the scatter plots of all the possible combination between the columns

```python
# The combination of same column over itself plots a count plot
sns.pairplot(iris)
```

```python
# creates an instance of an empty pair-plot called as Pair Grid
g = sns.PairGrid(iris)

# mapping the diagonal of the pairplot as a Distribution Plot
g.map_diag(sns.distplot)

# And all the plots above the diagonal as scatter plots
g.map_upper(plt.scatter)
```

```python
# loading the 'tips' dataset
tips = sns.load_dataset('tips')
```

```python
# Displaying the head(default = 5) of the 'tips' data
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col = 'time', row = 'smoker')
g.map(sns.distplot, 'total_bill')
```
